```python
import pandas as pd
import numpy as np
credits = pd.read_csv("tmdb_5000_credits.csv")
movies = pd.read_csv("tmdb_5000_movies.csv")
credits.head()
movies.head()
```

```python
print("Credits:",credits.shape)
print("Movies Dataframe:",movies.shape)
```

```python
credits_column_renamed = credits.rename(index=str, columns={"movie_id": "id"})
movies_merge = movies.merge(credits_column_renamed, on='id')
print(movies_merge.head())
```

```python
movies_cleaned = movies_merge.drop(columns=['homepage', 'title_x', 'title_y', 'status','production_countries'])
print(movies_cleaned.head())
print(movies_cleaned.info())
print(movies_cleaned.head(1)['overview'])
```

```python
!pip install scikit-learn

from sklearn.feature_extraction.text import TfidfVectorizer #   CONVERTS Data into Numerical feature vectors.
tfv = TfidfVectorizer(min_df=3,  max_features=None,
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

#    Breakdown of Parameters
# 
#   Parameter	Description
#   min_df=3	Ignores words that appear in fewer than 3 documents (removes rare words).
#   max_features=None	Uses all words instead of limiting to a set number of features.
#   strip_accents='unicode'	Converts accented characters (like é → e) for normalization.
#   analyzer='word'	Tokenizes based on words (rather than characters).
#   token_pattern=r'\w{1,}'	Defines how words are identified (matches words of at least 1 character).
#   ngram_range=(1, 3)	Extracts unigrams (1-word phrases), bigrams (2-word phrases), and trigrams (3-word phrases) for richer text representation.
#   stop_words='english'	Removes common English stop words (e.g., "the", "is", "and").
```

```python
# Fill NaN values in 'overview' column with an empty string
movies_cleaned['overview'] = movies_cleaned['overview'].fillna('')

tfv_matrix = tfv.fit_transform(movies_cleaned['overview'])
print(tfv_matrix)
print(tfv_matrix.shape)
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
print(sig[0])
```
